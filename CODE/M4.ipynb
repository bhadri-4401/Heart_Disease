### Catboost

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
from catboost import CatBoostClassifier
```

```python
data=pd.read_csv('heart.csv')
```

```python
data=data.dropna()
```

### Data Set

```python
data.head()
```

### Selecting The Dataset Parameters

```python
x=features=data.drop('output',axis=1)
y=target=data['output']
```

### Importing train_test_split from model_selection

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import explained_variance_score
```

### Splitting

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=32 )
```

### Importing tests 

```python
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix, classification_report
from math import sqrt
```

```python
Cat=CatBoostClassifier()
Cat.fit(x_train,y_train)
```

```python
predicted = Cat.predict(x_test)
```

Finding Accuracy

```python
accuracy = accuracy_score(y_test,predicted)
print('Accuracy of Catboost',accuracy*100)
```

Finding Clasiification Report

```python
cr = classification_report(y_test,predicted)
print('Classification report\n\n',cr)
```

Finding Confusion matrix

```python
cm = confusion_matrix(y_test,predicted)
print('Confusion matrix\n\n',cm)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,10))
plot_confusion_matrix(Cat, x_test, y_test, ax=ax)
plt.title('Confusion matrix of Catboost')
plt.show()
```
